```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Generate x values
x = np.linspace(-10, 10, 100)

# Calculate y values using the sigmoid function
y = sigmoid(x)

# Plot
plt.plot(x, y)
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.grid(True)
plt.show()
```

```python
def relu(x):
    return np.maximum(0 , x)

x = np.linspace(-10 , 10 , 100)

y = relu(x)

plt.plot(x, y)
plt.title('Relu Function')
plt.xlabel('x')
plt.ylabel('Relu(x)')
plt.grid(True)
plt.show()
```

```python
def leaky_relu(x, alpha=0.1):
    return np.maximum(alpha * x, x)

# Generate x values
x = np.linspace(-10, 10, 100)

# Calculate y values using the Leaky ReLU function
y = leaky_relu(x)

# Plot
plt.plot(x, y)
plt.title('Leaky ReLU Function')
plt.xlabel('x')
plt.ylabel('Leaky ReLU(x)')
plt.grid(True)
plt.show()
```

```python
def tanh(x):
    return np.tanh(x)


x = np.linspace(-10, 10, 100)

# Calculate y values using the Leaky ReLU function
y = tanh(x)

# Plot
plt.plot(x, y)
plt.title('tanh Function')
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.grid(True)
plt.show()
```

```python
def softmax(x):
    exp_scores = np.exp(x)
    return exp_scores/np.sum(exp_scores)


x = np.linspace(-10, 10, 100)

y = softmax(x)

# Plot
plt.plot(x, y)
plt.title('softmax Function')
plt.xlabel('x')
plt.ylabel('softmax(x)')
plt.grid(True)
plt.show()

```

```python
def sigmoid_prime(x):
    return sigmoid(x) * (1 - sigmoid(x))

x = np.linspace(-10, 10, 100)

# Calculate y values using the Leaky ReLU function
y = sigmoid_prime(x)

# Plot
plt.plot(x, y)
plt.title('sigmoid_prime Function')
plt.xlabel('x')
plt.ylabel('sigmoid_prime(x)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

class activation_function:
    def __init__(self, method):
        self.method = method
    
    def sigmoid(self, input_num):
        return 1 / (1 + np.exp(-input_num))
    
    def d_sigmoid(self, input_num):
        sigmoid_val = self.sigmoid(input_num)
        return sigmoid_val * (1 - sigmoid_val)
    
    def tanh(self, input_num):
        exp_plus = np.exp(input_num)
        exp_minus = np.exp(-input_num)
        return (exp_plus - exp_minus) / (exp_plus + exp_minus)
    
    def d_tanh(self, input_num):
        tanh_val = self.tanh(input_num)
        return 1 - tanh_val ** 2
    
    def softmax(self, input_num):
        exp_vals = [np.exp(x) for x in input_num - np.max(input_num)]
        exp_sum = sum(exp_vals)
        return [exp_val / exp_sum for exp_val in exp_vals]
    
    def relu(self, input_num):
        return np.maximum(0, input_num)
    
    def d_relu(self, input_num):
        return np.where(input_num > 0, 1, 0)
    
    def leaky_relu(self, input_num, alpha=0.01):
        return np.where(input_num > 0, input_num, alpha * input_num)
    
    def d_leaky_relu(self, input_num, alpha=0.01):
        return np.where(input_num > 0, 1, alpha)
    
    def gelu(self, input_num):
        return 0.5 * input_num * (1 + self.tanh(0.79788456 * (input_num + 0.044715 * input_num ** 3)))
    
    def d_gelu(self, input_num):
        tanh_val = self.tanh(0.79788456 * (input_num + 0.044715 * input_num ** 3))
        return 0.5 * (1 + tanh_val + (0.79788456 * (1 - tanh_val ** 2) * (0.044715 * input_num ** 2 + 0.75 * input_num * (1 - tanh_val ** 2))))
    
# Create input values
x = np.linspace(-5, 5, 100)

# Initialize activation function object
act_func = activation_function(method=None)

# Plot each activation function
plt.figure(figsize=(12, 8))

plt.subplot(3, 3, 1)
plt.plot(x, act_func.sigmoid(x))
plt.title('Sigmoid')
plt.grid(True)

plt.subplot(3, 3, 2)
plt.plot(x, act_func.tanh(x))
plt.title('Tanh')
plt.grid(True)

plt.subplot(3, 3, 3)
plt.plot(x, act_func.relu(x))
plt.title('ReLU')
plt.grid(True)

plt.subplot(3, 3, 4)
plt.plot(x, act_func.leaky_relu(x))
plt.title('Leaky ReLU')
plt.grid(True)

plt.subplot(3, 3, 5)
plt.plot(x, act_func.softmax(x))
plt.title('Softmax')
plt.grid(True)

plt.subplot(3, 3, 6)
plt.plot(x, act_func.gelu(x))
plt.title('GELU')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

class activation_function:
    def __init__(self, method):
        self.method = method
    
    def sigmoid(self, input_num):
        return 1 / (1 + np.exp(-input_num))
    
    def d_sigmoid(self, input_num):
        sigmoid_val = self.sigmoid(input_num)
        return sigmoid_val * (1 - sigmoid_val)
    
    def tanh(self, input_num):
        exp_plus = np.exp(input_num)
        exp_minus = np.exp(-input_num)
        return (exp_plus - exp_minus) / (exp_plus + exp_minus)
    
    def d_tanh(self, input_num):
        tanh_val = self.tanh(input_num)
        return 1 - tanh_val ** 2
    
    def softmax(self, input_num):
        exp_vals = np.exp(input_num - np.max(input_num))
        exp_sum = np.sum(exp_vals)
        return exp_vals / exp_sum
    
    def d_softmax(self, input_num):
        softmax_val = self.softmax(input_num)
        return softmax_val * (1 - softmax_val)
    
    def relu(self, input_num):
        return np.maximum(0, input_num)
    
    def d_relu(self, input_num):
        return np.where(input_num > 0, 1, 0)
    
    def leaky_relu(self, input_num, alpha=0.01):
        return np.where(input_num > 0, 1, alpha)
    
    def d_leaky_relu(self, input_num, alpha=0.01):
        return np.where(input_num > 0, 1, alpha)
    
    def gelu(self, input_num):
        return 0.5 * input_num * (1 + self.tanh(0.79788456 * (input_num + 0.044715 * input_num ** 3)))
    
    def d_gelu(self, input_num):
        tanh_val = self.tanh(0.79788456 * (input_num + 0.044715 * input_num ** 3))
        return 0.5 * (1 + tanh_val + (0.79788456 * (1 - tanh_val ** 2) * (0.044715 * input_num ** 2 + 0.75 * input_num * (1 - tanh_val ** 2))))
    
# Create input values
x = np.linspace(-5, 5, 100)

# Initialize activation function object
act_func = activation_function(method=None)

# Plot each activation function's derivative
plt.figure(figsize=(12, 8))

plt.subplot(3, 3, 1)
plt.plot(x, act_func.d_sigmoid(x))
plt.title('Sigmoid Derivative')
plt.grid(True)

plt.subplot(3, 3, 2)
plt.plot(x, act_func.d_tanh(x))
plt.title('Tanh Derivative')
plt.grid(True)

plt.subplot(3, 3, 3)
plt.plot(x, act_func.d_relu(x))
plt.title('ReLU Derivative')
plt.grid(True)

plt.subplot(3, 3, 4)
plt.plot(x, act_func.d_leaky_relu(x))
plt.title('Leaky ReLU Derivative')
plt.grid(True)

plt.subplot(3, 3, 5)
# Since softmax derivative w.r.t. input is not a commonly used operation, plotting it for visualization purposes only
plt.plot(x, act_func.d_softmax(x))
plt.title('Softmax Derivative')
plt.grid(True)

plt.subplot(3, 3, 6)
plt.plot(x, act_func.d_gelu(x))
plt.title('GELU Derivative')
plt.grid(True)

plt.tight_layout()
plt.show()
```
